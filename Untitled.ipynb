Импортируем черепашку. Создаем объект t с типом черепашка. Рисуем линию, длинной 100 пикселей. Черепашка вначале программы всегда смотрит вправо.

```python
import turtle
t = turtle.Turtle()
t.forward(100)
```

Поворачиваем налево на 60 градусов.

```python
import turtle
t = turtle.Turtle()
t.forward(100)
t.left(60)
t.forward(100)
```

Поворачивая на 90 градусов, 4 раза рисуем квадрат.

```python
import turtle
t = turtle.Turtle()
t.speed(10)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
```

Чтобы сократить код, используем цикл for. Повторив его 4 раза рисуем квадрат. Ускоряем рисование с помощбю функции speed.

```python
import turtle
t = turtle.Turtle()
t.speed(10)
for i in range(4):
    t.forward(100)
    t.left(90)
```

Создаем функцию sq, которая рисует квадрат. Вызываем её.

```python
import turtle
t = turtle.Turtle()
def sq():
    for i in range(4):
        t.forward(100)
        t.left(90)
sq()
```

Меняя параметр функции sq меняем размер стороны квадрата.

```python
import turtle
t = turtle.Turtle()
def sq(a):
    for i in range(4):
        t.forward(a)
        t.left(90)
sq(100)
sq(50)
```

Создав переменную length, устанавливаем длину стороны квадрата. Увеличивая её в цикле, рисуем десять квадратов разных размеров.

```python
import turtle
t = turtle.Turtle()
def sq(a):
    for i in range(4):
        t.forward(a)
        t.left(90)
length = 40
for i in range(10):
    sq(length)
    length=length+20
```

Поворачивая черепашку в цикле, создаем узор.

```python
import turtle
t = turtle.Turtle()
t.speed(100)
def sq(a):
    for i in range(4):
        t.forward(a)
        t.left(90)
length = 20
for i in range(36):
    sq(length)
    t.right(10)
    length=length+5
```

Добавил список colors, включаем в него цвета радуги. После в цикле используем каждый цвет по очереди. 

```python
import turtle
t = turtle.Turtle()
t.speed(100)
colors = ['red', "orange", 'yellow', "green", "light blue", "blue", "purple"]
def sq(a):
    for i in range(4):
        t.forward(a)
        t.left(90)
length = 20
for i in range(36):
    t.color(colors[i%7])
    sq(length)
    t.right(10)
    length=length+5
```
